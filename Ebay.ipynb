```python
import undetected_chromedriver as uc
from selenium.webdriver import ActionChains
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import time
```

```python
def getLongDecsription(url):
    item_descr_url = 'https://vi.vipr.ebaydesc.com/ws/eBayISAPI.dll?item={item_id}'
    item_id = url.split('?')[0].split('/')[-1]
    soup = BeautifulSoup(requests.get(item_descr_url.format(item_id=item_id)).content, 'html.parser')
    return soup.get_text(strip=True, separator='\n')

def Ebay(url):
    options = webdriver.ChromeOptions() 
    options.headless = True
    driver = uc.Chrome(options=options)
#     driver = uc.Chrome()
    driver.get(url)
    driver.maximize_window()
    print("Loaded")
    time.sleep(2)
    title = driver.find_element(By.CLASS_NAME, "x-item-title__mainTitle").text
    price = driver.find_element(By.CLASS_NAME, 'x-bin-price__content').text.split("$")[-1]
    images_divs = driver.find_elements(By.CLASS_NAME, "ux-image-carousel-item")
    images = []
    for img in images_divs:
        img_url = img.find_element(By.TAG_NAME, "img").get_attribute("src")
        if img_url is None:
            img_url = img.find_element(By.TAG_NAME, "img").get_attribute("data-src")
        images.append(img_url)
    feature_image = images[0]
    long_description = getLongDecsription(url)
    driver.quit()
    return {
        "title": title,
        "images": images,
        "featured_image": feature_image,
        "long_description": long_description,
        "price": price
    }
```
